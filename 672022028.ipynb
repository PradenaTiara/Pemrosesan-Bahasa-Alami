### **Ambil Kolom full_text & Tampilkan isi dari kolom bernama full_text**




```python
# Ambil hanya kolom 'full_text' dan hilangkan NaN
df = df[['full_text']].dropna().astype(str)

# Ubah Tampilkan
styled_df = df.head(10).style.set_properties(
    **{'background-color': '#ffc0cb',
       'color': 'black',
       'border': '1px solid black',
       'text-align': 'left'}
).set_table_styles(
    [{'selector': 'th', 'props': [('background-color', '#ffb6c1'), ('color', 'black'), ('text-align', 'center')]}]
)

styled_df
```

###**Proses Cleaning**
Hapus simbol dan karakter yang tidak diinginkan



```python
def clean_text_symbols(text):
    # Hapus simbol dan karakter selain huruf, angka, dan spasi
    return re.sub(r'[^a-zA-Z0-9\s]', '', str(text))

# Terapkan ke kolom 'full_text'
df['cleaned'] = df['full_text'].apply(clean_text_symbols)
df[['full_text', 'cleaned']].head()
```

### **Proses Case_Folding**

Ubah semua huruf besar menjadi huruf kecil

```python
df['case_folded'] = df['cleaned'].apply(lambda x: x.lower())
df[['cleaned', 'case_folded']].head()
```

```python
df['tokenized'] = df['case_folded'].apply(lambda x: x.split())
df[['case_folded', 'tokenized']].head()
```
